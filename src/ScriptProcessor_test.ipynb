```python
from sagemaker.processing import ScriptProcessor
from sagemaker.processing import ProcessingInput, ProcessingOutput
```

```python
myImage = "201513274994.dkr.ecr.us-east-2.amazonaws.com/hack2020:latest"
```

```python
import sagemaker
role = sagemaker.get_execution_role()
sess = sagemaker.Session()
bucket = sess.default_bucket()
```

```python
#print(bucket)
```

```python
#print(role)
```

```python
script_processor = ScriptProcessor(image_uri=myImage,
                role = role,
                instance_count = 1,instance_type='ml.t3.medium',
                command = ["python", "-v"])
```

```python
script_processor.run(code='s3://bbva-hack-2020/inputs/reconstruct.py',
    inputs=[ProcessingInput(
        source='s3://bbva-hack-2020/inputs/sample_10K.parquet',
        destination='/opt/ml/processing/input_data')],
        outputs=[ProcessingOutput(source='/opt/ml/processing/processed_data')])
```
